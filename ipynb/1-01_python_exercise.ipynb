## ③if文

if文の処理を記述しましょう

```
for i in range(10):
    if i > 0: #条件1
        print(i, "is lager than zero")
    elif i == 0: #条件2
        print(i, "is zero")
    else: #いずれの場合でもない条件
        print(i, "is less than zero")
```

## ④関数

関数でインプット(引数)とアウトプット(return)を変えて記述していましょう
*   サブルーチン処理(定型処理)を定義できます
*   インプット(args; 引数)とアウトプット(return; 戻り値)を定義し、プログラミング効率的に記述できます



```
def sum1(x, y):
    ans = x + y
    return ans

print(sum1(3, 5))
```

機械学習のテキストでしばしばlambda式で記述されている関数もあるため紹介しておきます

```
sum2 = lambda x,y: x+y
print(sum2(3,5))
```

## ⑤演習

*   リスト内の数字の内、正の整数のみ足す関数を定義し実行してみましょう
*   負の数を足す関数などアレンジしてみましょう




```
def pos_sum(num_list):
  a = 0 #初期値を定義
  for i in num_list:
    if i > 0:
      a = a + i
  return a
```

```
int_list = [1, -3, 4, 10, -8] #リストで複数の数字を定義
```

```
pos_sum(int_list)
```
