```
""""
#1
Write a code generating the following vector B from a given vector A. 

A:
array([9, 4, 5, 1, 8, 6])


B:
array([1, 4, 5, 6, 8, 9])
""""
A=np.array([9, 4, 5, 1, 8, 6])

def test(A):
    return np.sort(A)

print(test(A))
```

```
""""
#2
Write a code generating the following list of ndarray B from a given ndarray A. 
 
A:
array([[2, 5, 2, 6, 3, 2, 5, 1, 9, 4, 5, 6],
       [5, 8, 4, 9, 3, 1, 7, 9, 8, 3, 5, 8],
       [5, 1, 2, 7, 2, 1, 1, 2, 5, 4, 9, 3],
       [7, 7, 7, 9, 9, 9, 2, 5, 6, 4, 3, 8]])


B:
[array([[2, 5, 2, 6, 3, 2, 5, 1, 9, 4, 5, 6],
       [5, 8, 4, 9, 3, 1, 7, 9, 8, 3, 5, 8]]),
 array([[5, 1, 2, 7, 2, 1, 1, 2, 5, 4, 9, 3],
       [7, 7, 7, 9, 9, 9, 2, 5, 6, 4, 3, 8]])]


A's shape: (4, 12)
""""
A=np.array([[2, 5, 2, 6, 3, 2, 5, 1, 9, 4, 5, 6],
       [5, 8, 4, 9, 3, 1, 7, 9, 8, 3, 5, 8],
       [5, 1, 2, 7, 2, 1, 1, 2, 5, 4, 9, 3],
       [7, 7, 7, 9, 9, 9, 2, 5, 6, 4, 3, 8]])

def test(A):
    return np.split(A,[2])

print(test(A))
```

```
"""
#4
Write a code generating the following list of ndarray B from a given ndarray A. 
 
A:
array([[2, 5, 1, 4, 2, 4],
       [3, 9, 2, 3, 7, 3],
       [2, 8, 1, 8, 5, 3]])


B:
[array([[2, 5, 1],
       [3, 9, 2],
       [2, 8, 1]]),
 array([[4, 2, 4],
       [3, 7, 3],
       [8, 5, 3]])]


A's shape: (3, 6)
"""
A=np.array([[2, 5, 1, 4, 2, 4],
       [3, 9, 2, 3, 7, 3],
       [2, 8, 1, 8, 5, 3]])

def test(A):
    return np.hsplit(A,[3])

print(test(A))
```

```
"""
#5
 Write a code generating the following ndarray B from a given ndarray A. 

A:
array([[1, 5, 7, 7],
       [4, 9, 9, 3],
       [8, 8, 5, 6]])


B:
array([[False, False,  True,  True],
       [False,  True,  True, False],
       [ True,  True, False,  True]])


A's shape: (3, 4), B's shape: (3, 4) 

Hint: 5
"""

A=np.array([[1, 5, 7, 7],
       [4, 9, 9, 3],
       [8, 8, 5, 6]])

def test(A):
    return (A > 5)

print(test(A))
```

```
"""
#6
Write a code generating the following ndarray C from given ndarray A and ndarray B. 
 
A:
array([[3, 5, 0, 6, 2],
       [0, 7, 3, 5, 7],
       [1, 4, 4, 7, 5],
       [7, 0, 6, 3, 3]])


B:
array([[4, 7, 3, 4, 7],
       [3, 0, 4, 5, 5],
       [7, 3, 3, 5, 8],
       [7, 5, 6, 3, 4]])


C:
array([[3, 5, 0, 6, 2, 4, 7, 3, 4, 7],
       [0, 7, 3, 5, 7, 3, 0, 4, 5, 5],
       [1, 4, 4, 7, 5, 7, 3, 3, 5, 8],
       [7, 0, 6, 3, 3, 7, 5, 6, 3, 4]])


A's shape: (4, 5)    B's shape: (4, 5)    C's shape: (4, 10)
"""
A=np.array([[3, 5, 0, 6, 2],
       [0, 7, 3, 5, 7],
       [1, 4, 4, 7, 5],
       [7, 0, 6, 3, 3]])


B=np.array([[4, 7, 3, 4, 7],
       [3, 0, 4, 5, 5],
       [7, 3, 3, 5, 8],
       [7, 5, 6, 3, 4]])

def test(A,B):
    return np.hstack((A,B))

print(test(A,B))
```

```
"""
#7
Write a code generating the following ndarray C from given ndarray A and ndarray B. 
 
A:
array([[2, 3, 2],
       [2, 3, 3],
       [2, 1, 1]])


B:
array([[3, 3, 3],
       [2, 2, 1],
       [3, 1, 1]])


C:
array([[18, 14, 11],
       [21, 15, 12],
       [11,  9,  8]])


A's shape: (3, 3)    B's shape: (3, 3)    C's shape: (3, 3)
"""
A=np.array([[2, 3, 2],
       [2, 3, 3],
       [2, 1, 1]])


B=np.array([[3, 3, 3],
       [2, 2, 1],
       [3, 1, 1]])

def test(A, B):
    return np.dot(A,B)

print(test(A,B))
```

```
"""
#8
 Write a code generating the following ndarray B from a given ndarray A. 

A:
array([[ 3, 18,  6,  4,  3, 18,  7],
       [13, 11, 11, 19, 12, 19,  9],
       [ 2, 14,  4,  2, 18, 17, 13],
       [ 2, 15,  8,  3, 12, 13,  5],
       [ 7, 15, 13,  4, 16, 16, 10],
       [ 3, 13,  3,  4,  8, 16, 11],
       [12,  1, 14, 14,  8, 19, 13]])


B:
array([[3, 0, 6, 4, 3, 0, 7],
       [0, 0, 0, 0, 0, 0, 9],
       [2, 0, 4, 2, 0, 0, 0],
       [2, 0, 8, 3, 0, 0, 5],
       [7, 0, 0, 4, 0, 0, 0],
       [3, 0, 3, 4, 8, 0, 0],
       [0, 1, 0, 0, 8, 0, 0]])


A's shape: (7, 7), B's shape: (7, 7) 

Hint: 10
"""
A=np.array([[ 3, 18,  6,  4,  3, 18,  7],
       [13, 11, 11, 19, 12, 19,  9],
       [ 2, 14,  4,  2, 18, 17, 13],
       [ 2, 15,  8,  3, 12, 13,  5],
       [ 7, 15, 13,  4, 16, 16, 10],
       [ 3, 13,  3,  4,  8, 16, 11],
       [12,  1, 14, 14,  8, 19, 13]])

def test(A):
    return np.where(A>=10,0,A)

print(test(A))
```

```
"""
#9
Write a code generating the following ndarray B from a given ndarray A. 
 
A:
array([[0, 2, 5, 4, 5],
       [8, 2, 3, 8, 2],
       [3, 2, 4, 1, 8],
       [8, 0, 8, 7, 2],
       [0, 3, 8, 7, 3]])


B:
array([[16],
       [23],
       [18],
       [25],
       [21]])


A's shape: (5, 5)    B's shape: (5, 1)
"""
A=np.array([[0, 2, 5, 4, 5],
       [8, 2, 3, 8, 2],
       [3, 2, 4, 1, 8],
       [8, 0, 8, 7, 2],
       [0, 3, 8, 7, 3]])

def test(A):
    return np.sum(A,axis=1).reshape(5,1)

print(test(A))  
```

```
"""
#14
Write a code generating the following ndarray B from a given ndarray A. 
 
A:
array([[3, 5, 0, 6, 4, 2],
       [1, 6, 2, 5, 3, 3],
       [1, 0, 0, 2, 3, 0],
       [6, 5, 6, 4, 8, 4],
       [2, 1, 7, 4, 5, 2],
       [2, 6, 4, 6, 3, 6],
       [8, 7, 2, 0, 1, 8]])


B:
array([[8, 7, 7, 6, 8, 8]])


A's shape: (7, 6)    B's shape: (1, 6)
"""
A=np.array([[3, 5, 0, 6, 4, 2],
       [1, 6, 2, 5, 3, 3],
       [1, 0, 0, 2, 3, 0],
       [6, 5, 6, 4, 8, 4],
       [2, 1, 7, 4, 5, 2],
       [2, 6, 4, 6, 3, 6],
       [8, 7, 2, 0, 1, 8]])

def test(A):
    return np.max(A,axis=0)

print(test(A))
```
